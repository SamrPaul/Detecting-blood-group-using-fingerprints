```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['ResNet', 'AlexNet',  'LeNet']

# Epoch 5 Accuracy and Loss Data
accuracy = [0.8400, 0.1300, 0.9700]
val_accuracy = [0.7600, 0.1650, 0.4300]
loss = [0.1800, 2.3000, 0.0030]
val_loss = [0.7000, 2.3000, 5.0000]

# Bar width
bar_width = 0.2

# X-axis positions
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Plotting the bars
plt.figure(figsize=(10, 6))
bars1 = plt.bar(r1, accuracy, color='blue', width=bar_width, edgecolor='grey', label='Accuracy')
bars2 = plt.bar(r2, val_accuracy, color='green', width=bar_width, edgecolor='grey', label='Validation Accuracy')
bars3 = plt.bar(r3, loss, color='red', width=bar_width, edgecolor='grey', label='Loss')
bars4 = plt.bar(r4, val_loss, color='orange', width=bar_width, edgecolor='grey', label='Validation Loss')

# Add labels
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Accuracy/Loss', fontweight='bold')
plt.xticks([r + bar_width * 1.5 for r in range(len(models))], models)
plt.ylim(0, 6)

# Add values on top of bars with a diagonal angle
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 4), ha='center', va='bottom', rotation=45)

# Add legend
plt.legend()

# Display the plot
plt.title('Model Performance: Accuracy and Loss over 5 Epochs')
plt.show()
```
